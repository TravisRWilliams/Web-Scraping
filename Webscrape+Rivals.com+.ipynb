```python
import numpy as np
import pandas as pd
%pylab inline 
import bs4 as bs
import urllib.request
```

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

site= "https://auburn.rivals.com/commitments/football/2018"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site,headers=hdr)
page = urlopen(req)
soup = BeautifulSoup(page)
print(soup)
```

```python
pretty_soup=soup.prettify
print (pretty_soup)
```

```python
rivals_text=soup.get_text()
print (rivals_text)
```
